```julia
ENV["CUDA_VISIBLE_DEVICES"] = 1
```

```julia
using CNNVisualize, CuArrays, Metalhead
```

```julia
img = load_image("./images/cat_dog.png");
```

```julia
model = VGG19()
```

```julia
vbprop = VanillaBackprop(model.layers)
grads = vbprop(img)
pgrads, ngrads = positive_negative_saliency(grads[1][1]);
```

```julia
display("------------Vanilla Backprop Results------------")
display("-----------------Original Image-----------------")
display(load("./images/cat_dog.png"))
display("-----------Vanilla Backprop RGB Image-----------")
save_gradient_images(grads[1][1], "./results/vanilla_backprop_rgb_grads1_catdog.png")
display("--------Vanilla Backprop Grayscale Image--------")
save_grayscale_gradient(grads[1][1], "./results/vanilla_backprop_gray_grads1_catdog.png")
display("-------Positive Saliency Grayscale Image--------")
save_grayscale_gradient(pgrads, "./results/vanilla_backprop_posgray_grads1_catdog.png")
display("-------Negative Saliency Grayscale Image--------")
save_grayscale_gradient(ngrads, "./results/vanilla_backprop_neggray_grads1_catdog.png")
display("----------Positive Saliency RGB Image-----------")
save_gradient_images(pgrads, "./results/vanilla_backprop_posrgb_grads1_catdog.png")
display("----------Negative Saliency RGB Image-----------")
save_gradient_images(ngrads, "./results/vanilla_backprop_negrgb_grads1_catdog.png")
```

```julia
gbprop = GuidedBackprop(model.layers)
grads = gbprop(img)
pgrads, ngrads = positive_negative_saliency(grads[1][1]);
```

```julia
display("-------------Guided Backprop Results------------")
display("-----------------Original Image-----------------")
display(load("./images/cat_dog.png"))
display("-----------Guided Backprop RGB Image-----------")
save_gradient_images(grads[1][1], "./results/guided_backprop_rgb_grads1_catdog.png")
display("--------Guided Backprop Grayscale Image--------")
save_grayscale_gradient(grads[1][1], "./results/guided_backprop_gray_grads1_catdog.png")
display("-------Positive Saliency Grayscale Image--------")
save_grayscale_gradient(pgrads, "./results/guided_backprop_posgray_grads1_catdog.png")
display("-------Negative Saliency Grayscale Image--------")
save_grayscale_gradient(ngrads, "./results/guided_backprop_neggray_grads1_catdog.png")
display("----------Positive Saliency RGB Image-----------")
save_gradient_images(pgrads, "./results/guided_backprop_posrgb_grads1_catdog.png")
display("----------Negative Saliency RGB Image-----------")
save_gradient_images(ngrads, "./results/guided_backprop_negrgb_grads1_catdog.png")
```
